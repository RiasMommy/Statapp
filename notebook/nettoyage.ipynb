```python
import pandas as pd

# Adapter le chemin si besoin
file_path = "/home/onyxia/Statapp/data/rawdata.csv"

df = pd.read_csv(file_path)

# Vérification rapide
df.shape
```

```python
split = df["league"].str.split(",", expand=True)

split.head(15)
```

```python
# Vérifier la forme des lignes valides
valid = split[split[1].notna()]

valid.shape

valid.head(10)
```

```python

valid[6].value_counts().head(20)
```

```python
valid[5].value_counts().head(20)
```

```python
valid["pos"] = valid[6].str.replace('"', '', regex=False)
```

```python
valid["pos"].value_counts()
```

```python
valid = valid.drop(columns=[5, 6])
```

```python
valid["pos"].value_counts()
```

```python
valid = valid.rename(columns={
    0: "league",
    1: "season",
    2: "team",
    3: "player",
    4: "nation",
    7: "age",
    8: "born_year",
    9: "minutes"
})
```

```python
valid[["league", "season", "team", "player", "nation", "pos", "age", "born_year", "minutes"]].head(10)
```

```python
import pandas as pd

valid["season"]    = pd.to_numeric(valid["season"], errors="coerce")
valid["age"]       = pd.to_numeric(valid["age"], errors="coerce")
valid["born_year"] = pd.to_numeric(valid["born_year"], errors="coerce")
valid["minutes"]   = pd.to_numeric(valid["minutes"], errors="coerce")


```
