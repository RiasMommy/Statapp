



# Rapport de nettoyage — Projet StatApp

## 1. Fichier utilisé

Les données de départ viennent du fichier :
`data/rawdata.csv`

L’objectif était de rendre ce fichier exploitable pour les statistiques descriptives et les modèles.



## 2. Problèmes identifiés

En important le fichier brut, plusieurs points posaient problème :

* plusieurs colonnes étaient numérotées (10 à 44) sans nom,
* Source principale des valeurs manquantes : un mauvais merge.
Certaines lignes provenaient clairement d’une fusion incorrecte :
→ colonnes décalées
→ données alignées n’importe comment
→ lignes entières sans joueur (player = NaN) et avec toutes les statistiques vides.
Ces lignes ne représentaient aucun joueur réel
* des colonnes essentielles (`player`, `team`, `nation`, `age`, `minutes`) étaient parfois vides,
* une colonne (28) contenait des valeurs incohérentes (ordres de grandeur absurdes),
* beaucoup de données numériques étaient enregistrées comme du texte,
* deux colonnes étaient en doublon.


## 3. Nettoyage effectué

### 3.1 Suppression des lignes invalides

J’ai retiré :

* les lignes issues d’un mauvais merge (colonnes décalées, données illisibles),
* les lignes où les informations de base du joueur n’étaient pas présentes.

Ces lignes ne pouvaient pas être utilisées pour une analyse correcte.

### 3.2 Suppression d’un doublon

Deux colonnes contenaient exactement les mêmes valeurs.
J’en ai supprimé une pour éviter la redondance.

### 3.3 Suppression d’une colonne incohérente

La colonne 28 contenait des valeurs allant de 25 000 à 150 millions.
Aucune de ces valeurs ne correspond à une statistique footballistique normale.
Elle a donc été supprimée.

### 3.4 Renommage des colonnes

Les colonnes numérotées (10 → 44) ont été renommées à partir des intitulés officiels FBref
(exemple : `gls`, `ast`, `xg`, `xag`, `gls_per90`, `touches`, etc.).

Cela rend le dataset lisible et cohérent.

### 3.5 Conversion des types

Les colonnes contenant des chiffres étaient en réalité des chaînes de caractères.
Je les ai converties en `float64` pour pouvoir faire des calculs et des graphiques.



## 4. Fichiers produits

* `data/processed/players_clean.csv` : dataset propre, prêt pour l’analyse
* `data/processed/data_dictionary.csv` : description de chaque variable



## 5. Résultat

Le fichier final est maintenant propre, cohérent et exploitable pour :

* les statistiques descriptives,
* les visualisations,
* la modélisation.

Il remplace complètement le fichier brut initial.
